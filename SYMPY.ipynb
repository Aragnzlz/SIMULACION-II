```python
S = sp.Symbol('S')
t = sp.Symbol('t')
c = sp.Symbol('c')
```

```python
((t+4)**2).expand() #Desarrolla el polinomio
```

```python
sp.plotting.plot(t*t)
```

```python
sp.plotting.plot(5*sp.exp(0.4*t))
```

```python
S = sp.Function('S')
```

```python
ode = sp.Eq(sp.Derivative(S(t), t) - c * S(t), 0) #define la ecuación
ode
```

```python
sol = sp.dsolve(ode, ics={S(0): 10000}) #Solucion de la ecuación
sol
```

```python
constants = {c: 1}
```

```python
sol.subs(constants)
```

```python
sol1 = (sol).subs(c,1)
```

```python
sp.plotting.plot(sol1.rhs)
```

```python
sp.plotting.plot(sol.rhs)
```
